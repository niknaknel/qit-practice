```python
import numpy as np
import matplotlib.pyplot as plt

etas = np.linspace(0, 2*np.pi, 100)
Fb = [(1 + np.cos(eta))/2 for eta in etas]
Fe = [(1 + np.sin(eta))/2 for eta in etas]

ylabels = ["0", "$\\frac{\\pi}{4}$", "$\\frac{\\pi}{2}$", "$\\frac{3\\pi}{4}$", "$\\pi$",
           "$\\frac{5\\pi}{4}$", "$\\frac{3\\pi}{2}$", "$\\frac{7\\pi}{4}$", "$2\\pi$"]
print(ylabels)

plt.figure(figsize=(6,3))
plt.plot(etas, Fb)
plt.plot(etas, Fe)
plt.ylabel('Fidelity')
plt.xlabel('$\\eta$')
plt.xticks(np.linspace(0, 2*np.pi, 9), ylabels)
plt.xlim(0, 2*np.pi)
plt.ylim(0, 1)
plt.vlines([np.pi/4, 5*np.pi/4], 0, 1, 'lightblue', '--')
plt.legend(['$F_B$', '$F_E$'], loc='lower right')
```
